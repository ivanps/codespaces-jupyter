# Ciclos for

Repiten una instrucción varias veces (hasta que existan elementos en una lista)

```python
# Impreme los elementos de una lista
numeros=[10,20,30,40,50,60]
for num in numeros:
    print(num)
```

```python
lista1=['Ana',23,'Catalina', [29,18]]
for nom in lista1:
  print(nom)
```

```python
nombres = ['Pedro', 'Eduardo', 'Antonio', 'Margarito']
for nom in nombres:
  print(f'!Hola {nom}!')
```

```python
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune']
for planet in planets:
    print(planet, end=' ') # print all on same line
```

```python
print(range(2,5))
```

```python
for n in range(2,5):
  print(n)
```

```python
n = 5
for k in range(1,11):
  print(f' 3 x {k} = {n*k}')
```

```python
for k in range(1,11):
  print('*'*k)
```

```python
squares = []
for n in range(10):
    squares.append(n**2)
squares
```

```python
# Usando list comprehension
squares = [n**2 for n in range(10)]
squares
```

```python
# Termina un ciclo cuando se da una condición
for i in range(6):
    if i == 3:
        break
    print(i)
```

```python
# Salta un caso bajo una condición
for i in range(6):
    if i in [1,4]:
        continue
    print(i)
```

```python
# Otro ejemplo usando list comprehension
lista2 = [ 8, 21, 40, 2, 45, 27, 15 ]
impares = [i for i in lista2 if i%2 == 1]
impares
```
