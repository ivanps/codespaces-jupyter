```python
# Abre el archivo de la url https://www.ine.mx/wp-content/uploads/2023/01/DatosAbiertos-derfe-pdln_edms_sexo_20230921.xlsx y lo guarda en la variable df 
import pandas as pd

df = pd.read_excel('https://www.ine.mx/wp-content/uploads/2023/01/DatosAbiertos-derfe-pdln_edms_sexo_20230921.xlsx')
```

```python
# Muestra los primeros renglones del DataFrame
df.head(10)
```

```python
#Filtra por CLAVE\nENTIDAD el caso 32 (Zacatecas), asigno el resulta a otro DataFrame e imprime los primeros renglones
df2 = df[df['CLAVE\nENTIDAD'] == 32]
print(df2.shape)
df2.head()
```

```python
# De la columna SECCION, elimina valor cero y reporta su longitud
df3 = df2[df2['SECCION']!=0]
print(df3.shape)
df3.head()
```

```python
# Descarga el archivo zip de la liga https://cartografia.ife.org.mx/tmp/zips/Csecc-32.zip y desempaca los archivo que contiene
import requests, zipfile, io
r = requests.get('https://cartografia.ife.org.mx/tmp/zips/Csecc-32.zip')
z = zipfile.ZipFile(io.BytesIO(r.content))
z.extractall()
```

```python
# Lista los archivo en z
z.namelist()
```

```python
# Copia el archivo excel en z en un DataFrame
df4 = pd.read_excel(z.open('Catalogo de Secciones.xls'))
print(df4.shape)
df4.head()
```

```python
type(df4)
```

```python
# Imprime los valores de df4['SECCION'] que no están en df3['SECCION']
df4.loc[~df4["SECCION"].isin(df3["SECCION"]), ['NOMBRE MUNICIPIO', 'SECCION']]
```

```python
# how many gaps are there in the sequence of df3['SECCION'] sorted
lsecc = sorted(list(df3['SECCION']))
N = len(lsecc)
i = 1
gaps = 0
while i<=N-1:
    if lsecc[i] != lsecc[i-1]+1:
        gaps += 1
    i += 1
print(gaps)
```
